```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv_dataset_preprocessed.csv')
label = df['label']
print(df['url'].duplicated().sum())
label_counts = df['label'].value_counts()
benign_total = df[df['label'] == "Benign"].value_counts()
print(benign_total)
print(label_counts)
url_word_counts = df['url'].astype(str).apply(lambda x: len(x.split()))

# Kategorisasi berdasarkan jumlah kata
count_total_below_50 = (url_word_counts < 50).sum()
count_total_below_90 = (url_word_counts < 90).sum()
count_total_between_50_100 = ((url_word_counts >= 50) & (url_word_counts < 100)).sum()
count_total_between_100_150 = ((url_word_counts >= 100) & (url_word_counts < 150)).sum()
count_total_between_150_200 = ((url_word_counts >= 150) & (url_word_counts < 200)).sum()
count_total_between_200_300 = ((url_word_counts >= 200) & (url_word_counts < 300)).sum()
count_total_more_than_300 = (url_word_counts >= 300).sum()

# Cetak hasil
print("\nDistribusi Jumlah Kata dalam Kolom 'url':")
print(f"Jumlah kata < 50: {count_total_below_50 / len(df) * 100:.2f}%")
print(f"Jumlah kata < 90: {count_total_below_90 / len(df) * 100:.2f}%")
print(f"50 <= Jumlah kata < 100: {count_total_between_50_100 / len(df) * 100:.2f}%")
print(f"100 <= Jumlah kata < 150: {count_total_between_100_150/ len(df) * 100:.2f}%")
print(f"150 <= Jumlah kata < 200: {count_total_between_150_200/ len(df) * 100:.2f}%")
print(f"200 <= Jumlah kata < 300: {count_total_between_200_300/ len(df) * 100:.2f}%")
print(f"Jumlah kata >= 300: {count_total_more_than_300/ len(df) * 100:.2f}%")

# print avg, min, and max non_benign length
non_benign = df[df['label'] != "Benign"]
non_benign_url_word_counts = non_benign['url'].astype(str).apply(lambda x: len(x.split()))
avg_non_benign_length = non_benign_url_word_counts.mean()
print(f"\nRata-rata panjang URL untuk label non-Benign: {avg_non_benign_length:.2f} kata")
print(f"Jumlah URL non-Benign: {len(non_benign)}")
print(f"Panjang URL non-Benign maksimum: {non_benign_url_word_counts.max()} kata")
print(f"Panjang URL non-Benign minimum: {non_benign_url_word_counts.min()} kata")
print(f"Panjang URL non benign dibawah 50: {len(non_benign[non_benign_url_word_counts < 50])} kata")
print(f"Panjang URL non benign dibawah 90: {len(non_benign[non_benign_url_word_counts < 90])} kata")
print(f"Panjang URL non benign antara 50-100: {len(non_benign[(non_benign_url_word_counts >= 50) & (non_benign_url_word_counts < 100)])} kata")
print(f"Panjang URL non benign antara 100-150: {len(non_benign[(non_benign_url_word_counts >= 100) & (non_benign_url_word_counts < 150)])} kata")
print(f"Panjang URL non benign antara 150-200: {len(non_benign[(non_benign_url_word_counts >= 150) & (non_benign_url_word_counts < 200)])} kata")
print(f"Panjang URL non benign antara 200-300: {len(non_benign[(non_benign_url_word_counts >= 200) & (non_benign_url_word_counts < 300)])} kata")
print(f"Panjang URL non benign diatas 300: {len(non_benign[non_benign_url_word_counts >= 300])} kata")


# print avg, min, and max benign lenght
benign = df[df['label'] == "Benign"]
benign_url_word_counts = benign['url'].astype(str).apply(lambda x: len(x.split()))
avg_benign_length = benign_url_word_counts.mean()
print(f"\nRata-rata panjang URL untuk label Benign: {avg_benign_length:.2f} kata")
print(f"Jumlah URL Benign: {len(benign)}")
print(f"Panjang URL Benign maksimum: {benign_url_word_counts.max()} kata")
print(f"Panjang URL Benign minimum: {benign_url_word_counts.min()} kata")
print(f"Panjang URL benign dibawah 50: {len(benign[benign_url_word_counts < 50])} kata")
print(f"Panjang URL benign dibawah 90: {len(benign[benign_url_word_counts < 90])} kata")
print(f"Panjang URL benign antara 50-100: {len(benign[(benign_url_word_counts >= 50) & (benign_url_word_counts < 100)])} kata")
print(f"Panjang URL benign antara 100-150: {len(benign[(benign_url_word_counts >= 100) & (benign_url_word_counts < 150)])} kata")
print(f"Panjang URL benign antara 150-200: {len(benign[(benign_url_word_counts >= 150) & (benign_url_word_counts < 200)])} kata")
print(f"Panjang URL benign antara 200-300: {len(benign[(benign_url_word_counts >= 200) & (benign_url_word_counts < 300)])} kata")
print(f"Panjang URL benign diatas 300: {len(benign[benign_url_word_counts >= 300])} kata")
```
